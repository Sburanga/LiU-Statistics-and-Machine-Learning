```python
from __future__ import division
from math import radians, cos, sin, asin, sqrt, exp
from datetime import datetime
from pyspark import SparkContext
```

```python
path = '/home/h4pz/Downloads/data/'
```

```python
sc = SparkContext("local[*]")
sc.defaultParallelism
```

```python
# Prediction point.
latitude = 58.410809
longitude = 15.621373
date = "2000-05-08"

# Kernel data.
h_distance = 80 ** 2
h_date = 30 ** 2
h_hour = 3 ** 2

# Times to predict.
times = [f"{hour:02d}:00:00" for hour in range(4, 24, 2)] + ['00:00:00']

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance beteween two points
    on the earth (specified in decimal degrees)
    """
    
    # Convert decimal degrees to radians.
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    # Haversine formula.
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * asin(sqrt(a))
    km = 6367 * c
    return km

def day_distance(date1, date2):
    """Calculates the distance in days between two dates."""
    
    date1 = datetime.strptime(date1, '%Y-%m-%d').date()
    date2 = datetime.strptime(date2, '%Y-%m-%d').date()
    dist = (date1 - date2).days
    
    return abs(dist)

def hour_distance(hour1, hour2):
    """Calculates the distance in hours between two times."""
    hour1 = datetime.strptime(hour1, '%H:%M:%S')
    hour2 = datetime.strptime(hour2, '%H:%M:%S')
    dist = (hour1 - hour2).total_seconds() / (60 ** 2)
    
    return abs(dist)


def sum_gauss_kernel(km_dist, day_dist, hour_dist):
    """Computes a gaussian kernel."""
    dist_gk = exp(- km_dist / h_distance)
    day_gk = exp(- day_dist / h_date)
    hour_gk = exp(- hour_dist / h_hour)
    
    return dist_gk * day_gk * hour_gk

# Preparing the stations RDD.
stations = sc.textFile(path + "stations.csv")
stations = stations.map(lambda line: line.split(";"))

# (station, km_distance).
stations = stations.map(lambda x: (x[0], haversine(longitude,
                                                   latitude,
                                                   float(x[4]),
                                                   float(x[3]))))

stations = stations.coalesce(1)
stations = dict(stations.collect())
stations = sc.broadcast(stations)

# Preparing the temperature RDD.
temp = sc.textFile(path + "temperature-readings.csv").sample(False, 0.01)
temp = temp.map(lambda x: x.split(";"))

# Removing older dates.
temp = temp.filter(lambda x: x[1] < date)

# (station, (date, hour, temperature)).
# ('102170', ('2013-11-01', '06:00:00', 6.8)).
temp = temp.map(lambda x: (x[0], (x[1], x[2], float(x[3]))))

# Mapping all distances except for hour distance.
# (km_distance, day_distance, hour, temperature).
temp = temp.map(lambda x: (stations.value[x[0]],
                           day_distance(x[1][0], date),
                           x[1][1],
                           x[1][2]))

# Calculating the predicted temp for each time
# and saving them.
y_hat = []
for time in times:
    # Getting the kernel values for each observation.
    # (kernel_value, temperature).
    kernels = temp.map(lambda x: (sum_gauss_kernel(x[0],
                                                   x[1],
                                                   hour_distance(time, x[2])),
                                  x[3]))

    # (kernel_value, kernel_value * temperature).
    kernels = kernels.map(lambda x: (x[0], x[0] * x[1]))

    kernels = kernels.reduce(lambda x, y: (x[0] + y[0], x[1] + y[1]))
    y_hat.append(kernels[1] / kernels[0])
```
